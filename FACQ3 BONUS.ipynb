```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
def compare(tickers,weight1,weight2,start,end):
    data=yf.download(tickers,start=start,end=end)[["Adj Close"]]
    ndata=data.pct_change(fill_method=None)
    ndata.dropna()
    cdata1=ndata.dot(weight1)
    cdata2=ndata.dot(weight2)
    portfolioreturn1=(1+cdata1).cumprod()
    portfolioreturn2=(1+cdata2).cumprod()

    plt.plot(portfolioreturn1,label="RETURN1")
    plt.plot(portfolioreturn2,label="RETURN2")
    plt.xlabel=("YEAR")
    plt.ylabel=("RETURN")
    plt.legend()
    plt.show()

    portfoliovalue1=portfolioreturn1.iloc[-1]*1000000
    portfoliovalue2=portfolioreturn2.iloc[-1]*1000000  

    print("YOUR RETURN FROM INVESTMENT1 IS =",portfoliovalue1)
    print("YOUR RETURN FROM INVESTMENT2 IS =",portfoliovalue2)

    if(portfoliovalue1>portfoliovalue2):
        print("INVESTMENT1 IS GREAT")

    else:
       print("INVESTMENT2 IS GREAT")
```

```python
tickers=["MRF.NS","RELIANCE.NS","TSLA","REGN","VRTX"]
weight1=np.array([0.1,0.2,0.3,0.2,0.2])
weight2=np.array([0.2,0.1,0.2,0.3,0.2])
start="2023-01-01"
end="2023-12-31"
compare(tickers,weight1,weight2,start,end)
```
