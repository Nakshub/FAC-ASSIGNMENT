```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
def research(tickers,weights,start,end):
    data=yf.download(tickers,start=start,end=end)[["Adj Close"]]
    print(data.head())
    ndata=data.pct_change(fill_method=None)
    plt.figure(figsize=(10, 5)) 
    plt.plot(ndata, label=tickers)
    plt.title("Daily Percentage Change for Individual Stocks")
    plt.xlabel("Date")
    plt.ylabel("Percentage Change")
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,5))
    portfolio1=ndata.dot(weights)
    portfolioreturn=(1+portfolio1).cumprod()
    plt.plot(portfolioreturn)
    plt.xlabel("Date")
    plt.ylabel("Portfolio Return")

    
    print("Total portfolio value =",portfolioreturn["2023-12-29"]*1000000)
```

```python
tickers=["MRF.NS","RELIANCE.NS","TSLA","REGN","VRTX"]
weights=np.array([0.2,0.2,0.2,0.3,0.1])
start="2023-01-01"
end="2023-12-31"
research(tickers,weights,start,end)
```

```python
tickers=["MRF.NS","RELIANCE.NS","TSLA","AAPL","MSFT"]
weights=np.array([0.2,0.2,0.2,0.3,0.1])
start="2023-01-01"
end="2023-12-31"
research(tickers,weights,start,end)
```
